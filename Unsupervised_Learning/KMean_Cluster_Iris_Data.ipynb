```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# sns.get_dataset_names()
```

```python
df=sns.load_dataset("iris")
df.head()
```

```python
df["species"].value_counts()
```

```python
df_train=df[["sepal_length","sepal_width"]]
# df_train
```

```python
sns.pairplot(df_train)
```

```python
df_train.info()
```

```python
min_max_sale=MinMaxScaler()
df_train_scaled=min_max_sale.fit_transform(df_train)
# df_train_scaled
```

```python
from sklearn.cluster import KMeans
```

```python
kmean_cluster_model=KMeans(n_clusters=3)
kmean_cluster_model.fit(df_train_scaled)
```

```python
y_pred=kmean_cluster_model.predict(df_train_scaled)
y_pred
```

```python
df["Cluster"]=pd.DataFrame(y_pred)
```

```python
df
```

```python
df[df["species"]=="versicolor"]

df.groupby("Cluster")["Cluster"].count()
```

```python
SSE=[]
for eachK in range(1,10):
    kmean_cluster_model=KMeans(n_clusters=eachK)
    kmean_cluster_model.fit(df_train_scaled)
    SSE.append(kmean_cluster_model.inertia_)

SSE
```

```python
plt.plot(range(1,10),SSE)
```
