```python
def check_age_gender(row):
    return_val=True
    gender=row["gender"]
    age=int(row["age"])
    age_group=row["group"]
#     print(age_group.startswith(gender))
    if(not(age_group.startswith(gender))):
#         print(gender,age_group)
        return_val= False
#         return False
    else:
        age_bin=age_group[1:len(age_group)]
        if(age_bin.endswith("-")):
            if(age>int(age_bin[0:len(age_bin)-1])):
                return_val= False
        elif(age_bin.endswith("+")):
            if(age<int(age_bin[0:len(age_bin)-1])):
                return_val= False
        else:
            age_bin_split=age_bin.split("-")
#             print("nitesh",age_bin_split[0],age_bin_split[1])
            if((age<int(age_bin_split[0])) | (age>int(age_bin_split[1]))):
#                 print("reachinf")
                return_val= False 
    return return_val         
                
```

```python
df_gender_age_train["age_group_gender_check"]=df_gender_age_train.apply(check_age_gender,axis=1)
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="niteshs")
```

```python
lon_lat_state_dict=dict()
def get_state_city_based_long_lat(row):
    
    coord=str(row["latitude"])+","+str(row["longitude"])
#     print(coord)
    if coord not in lon_lat_state_dict:
        location = geolocator.reverse(coord)
        address = location.raw['address']
        city = address.get('county', '')
        state = address.get('state', '')
        lon_lat_state_dict[coord]=city+":"+state
#         print("**********",lon_lat_state_dict,"**********")
#         return lon_lat_state_dict
```

```python
def set_state_based_long_lat(row):
    coord=str(row["latitude"])+","+str(row["longitude"])
    print(row.index())
    if pd.isnull(row["state"]):
        row["newState"]=lon_lat_state_dict[coord]
    else:
        print(row["state"])
    
```
