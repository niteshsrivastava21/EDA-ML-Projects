```python
def get_country_wise_categorised_dataframe(df,country):
    try:
        df.set_index("Country/Region",inplace=True)
        df = df.swapaxes("index", "columns")
        df.reset_index(inplace=True)
        df["index"] = df["index"].astype(np.datetime64)
        df.set_index("index", inplace=True)
        df = pd.DataFrame(df[country])
    except RuntimeError:
        print(RuntimeError)
    return df
```

```python
df_confirmed_india=get_country_wise_categorised_dataframe(df_confirmed_cases,"Italy")
```
